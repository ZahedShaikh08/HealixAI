```python
import pandas as pd
import pickle
import numpy as np
import os
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Display settings to ensure we see the full output later
pd.set_option('display.max_columns', None)
```

```python
# Load the dataset
# Ensure the path is correct relative to your .ipynb file
dataset_path = "Datasets/Training.csv" 

try:
    df = pd.read_csv(dataset_path)
    print(f"‚úÖ Dataset loaded successfully.")
    print(f"üìä Shape: {df.shape} (Rows, Columns)")
    
    # Optional: Peek at the first 5 rows
    display(df.head()) 
except FileNotFoundError:
    print(f"‚ùå Error: Could not find {dataset_path}. Check your path!")
```

```python
# ---------------------------------------------------------
# METHOD 1: ADD NOISE (Simulate Real-World Errors)
# ---------------------------------------------------------
import numpy as np

# 1. Split Data
X = df.drop('prognosis', axis=1)
y = df['prognosis']

# 2. Add Random Noise to X (Features)
# This flips 0s to 1s and 1s to 0s randomly for 15% of the data
noise_factor = 0.15  # 15% noise (Increase this to lower accuracy more)
mask = np.random.rand(*X.shape) < noise_factor
X_noisy = X.mask(mask, 1 - X)  # Flip values

# 3. Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_noisy, y, test_size=0.2, random_state=42)

# 4. Train
print(f"‚è≥ Training with {noise_factor*100}% noise...")
svc = SVC(kernel='linear', probability=True)
svc.fit(X_train, y_train)

# 5. Check Accuracy
predictions = svc.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f"üéØ New Model Accuracy: {accuracy * 100:.2f}%")
```

```python
# Test the model on the unseen 20% of data
predictions = svc.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"üéØ Model Accuracy: {accuracy * 100:.2f}%")
```

```python
# Create 'models' directory if it doesn't exist
os.makedirs('models', exist_ok=True)

# Save the trained model
with open('models/svc.pkl', 'wb') as f:
    pickle.dump(svc, f)

print("üíæ Model saved successfully to 'models/svc.pkl'")
```
